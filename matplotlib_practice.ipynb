```python
import pandas as pd
data=pd.read_csv("//content//Titanic Dataset.csv",delimiter=',')
data
```

```python
!pip install matplotlib
from matplotlib import pyplot as plt
import numpy as np
!pip install seaborn
```

```python
#bar:
data['survived'].value_counts().plot(kind='bar')

plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
data['age'].plot(kind='hist')

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter:
data.plot(kind='scatter', x='age', y='fare')

plt.title("Age vs Fare")
plt.show()
```

```python
# line plot (normal line)
x = data['pclass'].head(10)
y = data['fare'].head(10)

plt.plot(x, y)
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Line Plot Example")
plt.show()
```

```python
# line plot without lines:
x = data['pclass'].head(10)
y = data['fare'].head(10)

plt.plot(x, y, marker="o", linestyle="")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Without Line")
plt.show()
```

```python
# line plot with grid:
x = data['pclass']
y = data['fare']

plt.plot(x, y)
plt.grid(True)

plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Line with Grid")
plt.show()
```

```python
#dotted line:
x = data['pclass'].head(10)
y = data['fare'].head(10)

plt.plot(x, y, linestyle='dotted',color="r")

plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Dotted Line")
plt.show()
```

```python
# bar with horizontal:
x = data['pclass'].value_counts().index
y = data['pclass'].value_counts().values

plt.barh(x, y)

plt.xlabel("Count")
plt.ylabel("Passenger Class")
plt.title("Horizontal Bar")
plt.show()
```

```python
# subplot:
plt.subplot(1,2,1)
plt.plot(data['age'].head(10))
plt.title("Age")

plt.subplot(1,2,2)
plt.plot(data['fare'].head(10))
plt.title("Fare")

plt.show()
```

```python
# Legend:
plt.plot(data['age'].head(10), label="Age",color="r")
plt.plot(data['fare'].head(10), label="Fare",color='g')

plt.legend()

plt.title("Legend Example")
plt.show()
```

```python
# xticks:
x = data['pclass'].head(10)
y = data['fare'].head(10)

plt.plot(x,y)

plt.xticks([1,2,3])

plt.xlabel("Class",color='purple')
plt.ylabel("Fare",color='green')
plt.title("Xticks Example")

plt.show()
```

```python
#pie chart:
data = data['sex'].value_counts()

plt.pie(data, labels=data.index, autopct='%1.1f%%')

plt.title("Male vs Female")
plt.show()
```
